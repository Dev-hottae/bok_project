# 국고채(3년만기) 크롤링 후 데이터프레임 생성

```
from bs4 import BeautifulSoup
import requests
import json
import pandas as pd
```

```
# treasury bond
dic = {}
for i in range(1, 541): # 541까지 // 2020.07.23 기준
    res = requests.get("https://finance.naver.com/marketindex/interestDailyQuote.nhn?marketindexCd=IRR_GOVT03Y&page={}".format(i))
    bs = BeautifulSoup(res.text, "html.parser")

    for j in bs.select('div tbody tr'):
        date = j.select_one('td.date').text.strip()
        rate = j.select_one('td.num').text.strip()
        dic.setdefault(date, rate)
dic
```

```
with open("treasury_b.json","w") as file:
    json.dump(dic, file)
```

```
# 파일 확인하기
with open('treasury_b.json', "r") as json_file:
    json_data = json.load(json_file)
```

```
# DataFrame에 넣기
# treasury bond
dic1 = {}
for i in range(1, 300):
    res = requests.get("https://finance.naver.com/marketindex/interestDailyQuote.nhn?marketindexCd=IRR_GOVT03Y&page={}".format(i))
    bs = BeautifulSoup(res.text, "html.parser")

    for j in bs.select('div tbody tr'):
        date = j.select_one('td.date').text.strip()
        rate = j.select_one('td.num').text.strip()
        dic1.setdefault(date, rate)
dic1
```

```
# treasury 3year DataFrame화
df = pd.DataFrame(dic1, index=dic1.keys(), columns=['rate'])
df['rate']=dic1.values()
df
```
